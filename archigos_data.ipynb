# Archigos: 世界政治リーダーDB

```python
import pandas as pd
import codecs
```

## データファイルの読み込み

```python
with codecs.open("archigos.csv",mode ="r", encoding= "cp932",errors="ignore") as file:
    df1 = pd.read_table(file, delimiter="\t")
df1.head()
```

```python
df1.info()
```

```python
df2 = df1[df1['idacr'].str.contains('JPN')]
df2.head()
```

## 国名テーブルを読み込み横連結する

```python
# 国名テーブルの読み込み
df101 = pd.read_excel('country_table.xlsx',sheet_name='table')
df101.head()
```

```python
# 3桁国コードをキーにして横連結する
df3 = pd.merge(df2, df101, left_on='idacr', right_on='ccode3')
df3.head()
```
