# Archigos: 世界政治リーダーDB

```python
import pandas as pd
import codecs
```

## データファイルの読み込み

```python
with codecs.open("archigos.csv",mode ="r", encoding= "cp932",errors="ignore") as file:
    df1 = pd.read_table(file, delimiter="\t")
df1.head()
```

```python
df1.info()
```

```python
df1.columns
```

```python
# 特定の国に絞り込む
df51 = df1[df1['idacr'].str.contains('JPN')]
df51.sample(10)
```

## 列を絞り込む

```python
# 列を絞り込む
focus_cols = ['obsid', 'ccode', 'idacr', 'leader', 'startdate', 'enddate','dbpedia.uri']
df2 = df1.loc[:, focus_cols]
df2.head()
```

## 国名テーブルを読み込み横連結する

```python
# 国名テーブルの読み込み
df101 = pd.read_excel('country_table.xlsx',sheet_name='table')
df101.head()
```

```python
# 3桁国コードをキーにして横連結する
df3 = pd.merge(df2, df101, left_on='idacr', right_on='ccode3')
df3.sample(10)
```

```python
df3.columns
```

```python
# さらに列を絞り込む
focus_cols_2 = ['obsid', 'ccode', 'idacr', 'leader', 'startdate', 'enddate','dbpedia.uri', 'jch_country', 'ch_country']
df4 = df3.loc[:, focus_cols_2]
df4.sample(10)
```

```python
# 日付文字列を日付値に変換する
df4['start_date'] = pd.to_datetime(df4['startdate'])
df4['end_date'] = pd.to_datetime(df4['enddate'])
df4.drop(columns=['startdate','enddate'],inplace=True)
df4.sample(10)
```

## pickleファイルに保存する

```python
df4.to_pickle('archigos.pkl')
```
