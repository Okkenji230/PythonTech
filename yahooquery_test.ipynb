# yahooqueryライブラリの動作確認

```python
import numpy as np
import pandas as pd
from yahooquery import Ticker
import matplotlib.pyplot as plt
# matplotlibで日本語表示を可能にする
from matplotlib import rcParams
rcParams['font.family']='MS Gothic'
```

```python
# 米国会社コードを指定　日本の企業は'コード.T'
ticker = Ticker('9432.T')
```

```python
# View available modules on the Ticker class
cmodule = [i for i in Ticker.__dict__.keys() 
 if "_" not in i[:2] ]
print(cmodule)
```

```python
df_tmp = ticker.income_statement('q') # 四半期の損益計算書
df_tmp
```

```python
df_income = df_tmp[df_tmp['periodType'] == '3M']
```

```python
df_income.columns
```

```python
#損益計算書の指標の折れ線グラフ
def income_line(ycol1, ctitle1, ycol2, ctitle2):
    fig = plt.figure(figsize=(12,3))
    ax = fig.add_subplot(121)
    ax.plot(df_income['asOfDate'], df_income[ycol1], linewidth=3, color="red")
    ax.set_title(ctitle1)
    ax.set_xlabel("period")
    ax.set_ylabel("amount")
    ax = fig.add_subplot(122)
    ax.plot(df_income['asOfDate'], df_income[ycol2], linewidth=3, color="blue")
    ax.set_title(ctitle2)
    ax.set_xlabel("period")
    ax.set_ylabel("amount")
```

```python
# TotalRevenue=売上高 NetIncome=利益
income_line('TotalRevenue', '社：売上高', 'NetIncome', '社：純利益')
```
