## ライブラリのインポート

```python
import pandas as pd
import yfinance as yf
from statsmodels.tsa.stattools import coint
import numpy as np
from ta.volatility import BollingerBands
from pandas_datareader import data as pdr
```

## データ獲得関数を定義

```python
def yfdownload(ticker, start, end):
    yf.pdr_override()
    df_stock = pdr.get_data_yahoo(ticker, start, end)
    return df_stock
```

## ペアドレーディング戦略

```python
pt_price_1 = yfdownload(ticker='6146.T', start='2020-01-01', end='2023-12-21')['Close']
pt_price_2 = yfdownload(ticker='6857.T', start='2020-01-01', end='2023-12-21')['Close']

score, pvalue, _ = coint(pt_price_1, pt_price_2)
print('p-value=',pvalue)
ratios = pt_price_1 / pt_price_2

# ペアドレーディング戦略の実装
ratios.plot()
buy = ratios.copy()
sell = ratios.copy()
buy[ratios > ratios.mean()] = 0
sell[ratios < ratios.mean()] = 0
```

## モメンタム戦略

```python
ms_price = yfdownload(ticker='7203.T', start='2020-01-01', end='2023-12-21')
ms_price['return'] = np.log(ms_price['Close'] / ms_price['Close'].shift())
ms_price['momentum'] = ms_price['return'].rolling(window=3).mean() / ms_price['return'].rolling(window=3).std()
ms_price['signal'] = np.where(ms_price['momentum'] > 0, 1, -1)
ms_price['strategy'] = ms_price['signal'].shift() * ms_price['return']
ms_price.dropna(inplace=True)
ms_price['cumulative_return'] = (1 + ms_price['strategy']).cumprod()
```

```python
ms_price['cumulative_return'].plot()
```

## 逆張り戦略

```python
cs_price = yfdownload(ticker='USDJPY=X', start='2020-01-01', end='2023-12-21')
cs_price['return'] = np.log(cs_price['Close'] / cs_price['Close'].shift())
cs_price['z-score'] = (cs_price['return'] - cs_price['return'].rolling(window=21).mean()) / cs_price['return'].rolling(window=21).std()
cs_price['signal'] = np.where(cs_price['z-score'] > 0, -1, 1)
cs_price['strategy'] = cs_price['signal'].shift() * cs_price['return']
cs_price.dropna(inplace=True)
cs_price['cumulative_return'] = (1 + cs_price['strategy']).cumprod()
```

```python
cs_price['cumulative_return'].plot()
```

## ボリンジャーバンド戦略

```python
bb_price = yfdownload(ticker='9983.T', start='2020-01-01', end='2023-12-21')
indicator_bb = BollingerBands(close=bb_price["Close"], window=20, window_dev=2)
bb_price['bb_bbp'] = indicator_bb.bollinger_pband()
bb_price['bb_wband'] = indicator_bb.bollinger_wband()
bb_price['signal'] = np.where(bb_price['bb_bbp'] < 0, 1, np.where(bb_price['bb_bbp'] > 1, -1, 0))
bb_price['return'] = bb_price['Close'].pct_change()
bb_price['strategy'] = bb_price['signal'].shift() * bb_price['return']
bb_price.dropna(inplace=True)
bb_price['cumulative_return'] = (1 + bb_price['strategy']).cumprod()
```

```python
bb_price['cumulative_return'].plot()
```
