# mplfinanceによるオシレーター分析（その2）

オシレーターが理論的な上限値(下限値)や経験的な上限域(下限域)に接近あるいは上側(下側)の基準値を超えたら買われすぎ(売られすぎ)シグナルと判断する。

順張りではオシレーターが基準値を超えた(割った)後基準値を割ったら(上回ったら)〔逆張りでは単に基準値を超えたら(割ったら)〕、買い(売り)を手じまい、新規に売り(買い)建てる。

ただし上昇(下降)の勢いが強いとオシレーターが買われすぎ(売られすぎ)の領域に張り付くため、逆張りは注意が必要。

上昇トレンドに入ってしばらく経過し安定的に上昇している場面でオシレーターが上限値を上回ったら逆張りのシグナルとしてもよい。

〔内容要確認〕下降(上昇)トレンドが長く続いた後など、トレンド転換の可能性がある場面で、買われすぎ(売られすぎ)シグナルが出た時の逆張りには注意が必要。価格が底入れして(天井をつけて)上昇(下降)トレンドに転じる場面では、買われすぎ(売られすぎ)シグナルが出ることが多い。

中心線よりも下(上)にあったオシレーターが中心線を上回れば(下回れば)トレンドが上昇(下降)に転じたと示唆されるので、順張りの買い(売り)シグナルとなる。

## MACD

【MACD解説】MACD = 短期MA - 中期MA

価格が上昇(下降)トレンドにあるとき、MACDはプラス(マイナス)圏で推移する。

価格の上昇トレンドが強まって(弱まって)いるとき、MACDは増加(減少)する。

価格の下降トレンドが強まって(弱まって)いるとき、MACDのマイナス幅が拡大(縮小)する。

MACDヒストグラムはMACDの変化が加速しているのか減速しているのかを表している。

(参考記事)
https://plainenglish.io/blog/plot-stock-chart-using-mplfinance-in-python-9286fc69689

```python
# Add MACD as subplot
# .ewm(span=n)のとき指数平滑係数α=2/(span+1) ただしspan>=1
def MACD(df, window_slow, window_fast, window_signal):
    macd = pd.DataFrame()
    macd['ema_slow'] = df['Close'].ewm(span=window_slow).mean()
    macd['ema_fast'] = df['Close'].ewm(span=window_fast).mean()
    macd['macd'] = macd['ema_slow'] - macd['ema_fast']
    macd['signal'] = macd['macd'].ewm(span=window_signal).mean()
    macd['diff'] = macd['macd'] - macd['signal']
    macd['bar_positive'] = macd['diff'].map(lambda x: x if x > 0 else 0)
    macd['bar_negative'] = macd['diff'].map(lambda x: x if x < 0 else 0)
    return macd

macd = MACD(df, 12, 26, 9)

macd_plot = [
    mpf.make_addplot((macd['macd']), color='#D2691E', panel=2, ylabel='MACD', secondary_y=False), #D2691Eはチョコレート色
    mpf.make_addplot((macd['signal']), color='#800080', panel=2, secondary_y=False), #800080は紫
    mpf.make_addplot((macd['bar_positive']), type='bar', color='#4dc790', panel=2),
    mpf.make_addplot((macd['bar_negative']), type='bar', color='#fd6b6c', panel=2),
]
```

カラーコード一覧：https://itsakura.com/html-color-codes

チョコレート色で、株価上昇時に先に反応して上向くのがMACD。紫色で後から上向くのがMACDシグナル(移動平均)。棒グラフは両者の差(MACDヒストグラム)

```python
# 表題と凡例
ch_title = 'Stock Price Chart for ' + ticker
line_titles = ['MACD','SIGNAL','BAR+','BAR-']
# チャート描画：指定可能なtypeは'ohlc'(default), 'candle', 'line', 'renko', 'pnf'
fig, axes = mpf.plot(df,type='candle',figsize =(16,9), style='yahoo',
         addplot=macd_plot,volume=True,mav=(5,25,75), returnfig=True)
# 表題と凡例の指定
axes[0].set_title(ch_title)
axes[4].legend(line_titles)
```

【MACD解説】MACD = 短期MA - 中期MA

価格が上昇(下降)トレンドにあるとき、MACDはプラス(マイナス)圏で推移する。

価格の上昇トレンドが強まって(弱まって)いるとき、MACDは増加(減少)する。

価格の下降トレンドが強まって(弱まって)いるとき、MACDのマイナス幅が拡大(縮小)する。

MACDヒストグラムはMACDの変化が加速しているのか減速しているのかを表している。

## オーサムオシレーターと加速減速オシレーター

【AO解説】価格が上昇(下降)トレンドにあるとき、AOはプラス圏(マイナス圏)にあり、AOが拡大(減少)するときに上昇(下降)が加速している。

AOがマイナスからプラスに転じる場面が買いシグナル、プラスからマイナスに転じる場面が売りシグナル。

ADO = AO - AOの移動平均　であり、AOの加速・減速の度合いを示す。

```python
# subplotとしてオーサムオシレーターを描画する
# 各種指標の計算関数
def OSC_ADO(df, wshort, wlong):
  osc_ado = pd.DataFrame()
  # 中値
  osc_ado['med'] = (df['High']+df['Low']) / 2
  # オーサムオシレーター
  osc_ado['ao'] = osc_ado['med'].rolling(wshort).mean() - osc_ado['med'].rolling(wlong).mean()
  # 加速減速オシレーター
  osc_ado['ado'] = osc_ado['ao'] - osc_ado['ao'].rolling(wshort).mean()
  return osc_ado
# 指標計算関数を実行して結果をデータフレームに格納する
osc_ado = OSC_ADO(df, 5, 34)
# 追加プロットのリストを作成する
osc_plot = [
    mpf.make_addplot((osc_ado['ao']), panel=2, ylabel='AO', secondary_y=False),
    mpf.make_addplot((osc_ado['ado']), panel=3, ylabel='ADO', secondary_y=False)
]
```

```python
# 表題と凡例
ch_title = 'Stock Price Chart for ' + ticker
line_titles1 = ['ao']
line_titles2 = ['ado']
# チャート描画：指定可能なtypeは'ohlc'(default), 'candle', 'line', 'renko', 'pnf'
fig, axes = mpf.plot(df,type='candle',figsize =(16,9), style='yahoo',
         addplot=osc_plot,volume=True,mav=(15,25,75), returnfig=True)
# 表題と凡例の指定
axes[0].set_title(ch_title)
axes[4].legend(line_titles1)
axes[6].legend(line_titles2)
```

【AO解説】価格が上昇(下降)トレンドにあるとき、AOはプラス圏(マイナス圏)にあり、AOが拡大(減少)するときに上昇(下降)が加速している。

AOがマイナスからプラスに転じる場面が買いシグナル、プラスからマイナスに転じる場面が売りシグナル。

ADO = AO - AOの移動平均　であり、AOの加速・減速の度合いを示す。

## マス指数 (MI)

【MI解説】MIは日中高低差の過去平均に対する割合。

MIの上昇(下降)が長く続いた後で天井(大底)の接近を想定する。

```python
# subplotとしてマス指数(MI)を描画する
# 各種指標の計算関数
def OSC_MI(df,wshort,wlong):
  osc_mi = pd.DataFrame()
  # 日中高低差
  osc_mi['range'] = df['High'] - df['Low']
  # マス指数
  osc_mi['rn'] = osc_mi['range'].ewm(span=wshort).mean()
  osc_mi['ln'] = osc_mi['rn'].ewm(span=wshort).mean()
  osc_mi['mi'] = (osc_mi['rn']/osc_mi['ln']).rolling(wlong).sum()
  return osc_mi

# 指標計算関数を実行して結果をデータフレームに格納する
osc_mi = OSC_MI(df,9,25)
# 追加プロットのリストを作成する
osc_plot1 = [
    mpf.make_addplot((osc_mi['mi']), panel=2, ylabel='MI', secondary_y=False)
]
```

```python
# 表題と凡例
ch_title = 'Stock Price Chart for ' + ticker
line_titles = ['MI']
# チャート描画：指定可能なtypeは'ohlc'(default), 'candle', 'line', 'renko', 'pnf'
fig, axes = mpf.plot(df,type='candle',figsize =(16,9), style='yahoo',
         addplot=osc_plot1,volume=True,mav=(15,25,75), returnfig=True)
# 表題と凡例の指定
axes[0].set_title(ch_title)
axes[4].legend(line_titles)
```

【MI解説】MIは日中高低差の過去平均に対する割合。

MIの上昇(下降)が長く続いた後で天井(大底)の接近を想定する。

## モメンタムと価格変化率

【ROC解説】ROCがマイナスからプラス(プラスからマイナス)に転じる時は、下降(上昇)トレンドから上昇(下降)トレンドに転じる時であり、買い(売り)シグナルとなる。

```python
# subplotとして価格変化率(ROC)を描画する
# 各種指標の計算関数
def OSC_ROC(df,window):
  osc_roc = pd.DataFrame()
  # モメンタム
  osc_roc['momentum'] = df['Close'] - df['Close'].shift(window)
  # ROC
  osc_roc['roc'] = (df['Close']-df['Close'].shift(window)) / df['Close'].shift(window) * 100
  return osc_roc

# 指標計算関数を実行して結果をデータフレームに格納する
osc_roc = OSC_ROC(df,25)
# 追加プロットのリストを作成する
osc_plot1 = [
    mpf.make_addplot((osc_roc['roc']), panel=2, ylabel='ROC', secondary_y=False)
]
```

```python
# 表題と凡例
ch_title = 'Stock Price Chart for ' + ticker
line_titles = ['ROC']
# チャート描画：指定可能なtypeは'ohlc'(default), 'candle', 'line', 'renko', 'pnf'
fig, axes = mpf.plot(df,type='candle',figsize =(16,9), style='yahoo',
         addplot=osc_plot1,volume=True,mav=(15,25,75), returnfig=True)
# 表題と凡例の指定
axes[0].set_title(ch_title)
axes[4].legend(line_titles)
```

【ROC解説】ROCがマイナスからプラス(プラスからマイナス)に転じる時は、下降(上昇)トレンドから上昇(下降)トレンドに転じる時であり、買い(売り)シグナルとなる。

## トリックス (TRIX)

【TRIX解説】価格の三重指数平滑移動平均の価格変化率。

TRIXがプラス(マイナス)なら上昇(下降)トレンドにある。TRIXが下降から上昇(上昇から下降)に転じたタイミングが買い(売り)となる。

価格が上昇(下降)トレンドにあるにもかかわらず、TRIXが下降(上昇)傾向となる〔ネガティブ(ポジティブ)・ダイバージェンス〕場合は天井(底入れ)をつける予兆とされる。

```python
# subplotとしてトリックス(TRIX)を描画する
# 各種指標の計算関数
def OSC_TRIX(df,window):
  osc_trix = pd.DataFrame()
  # 三重指数平滑移動平均
  osc_trix['ema1'] = df['Close'].ewm(span=window).mean()
  osc_trix['ema2'] = osc_trix['ema1'].ewm(span=window).mean()
  osc_trix['ema3'] = osc_trix['ema2'].ewm(span=window).mean()
  # トリックス
  osc_trix['trix'] = (osc_trix['ema3']-osc_trix['ema3'].shift(1)) / osc_trix['ema3'].shift(1) * 100
  return osc_trix

# 指標計算関数を実行して結果をデータフレームに格納する
osc_trix = OSC_TRIX(df,9)
# 追加プロットのリストを作成する
osc_plot1 = [
    mpf.make_addplot((osc_trix['trix']), panel=2, ylabel='TRIX', secondary_y=False)
]
```

```python
# 表題と凡例
ch_title = 'Stock Price Chart for ' + ticker
line_titles = ['TRIX']
# チャート描画：指定可能なtypeは'ohlc'(default), 'candle', 'line', 'renko', 'pnf'
fig, axes = mpf.plot(df,type='candle',figsize =(16,9), style='yahoo',
         addplot=osc_plot1,volume=True,mav=(15,25,75), returnfig=True)
# 表題と凡例の指定
axes[0].set_title(ch_title)
axes[4].legend(line_titles)
```

【TRIX解説】価格の三重指数平滑移動平均の価格変化率。

TRIXがプラス(マイナス)なら上昇(下降)トレンドにある。TRIXが下降から上昇(上昇から下降)に転じたタイミングが買い(売り)となる。

価格が上昇(下降)トレンドにあるにもかかわらず、TRIXが下降(上昇)傾向となる〔ネガティブ(ポジティブ)・ダイバージェンス〕場合は天井(底入れ)をつける予兆とされる。

## 相対力指数 (RSI)

【RSI解説】RSIが下方基準線(20-30%)を下回れば売られすぎ、上方基準線(70-80%)を上回れば買われすぎ。

RSIがボトムアウト後に下方基準線を上回る時に買い参入し、さらに上昇して上方基準線を上回る時に買いを手仕舞う。

```python
# subplotとして相対力指数(RSI)を描画する
# 各種指標の計算関数
def OSC_RSI(df,window):
  osc_rsi = pd.DataFrame()
  # 買い方圧力
  osc_rsi['buy_power'] = np.maximum((df['Close']-df['Close'].shift(1)),0).rolling(window).sum()
  # 売り方圧力
  osc_rsi['sell_power'] = (-1) * np.minimum((df['Close']-df['Close'].shift(1)),0).rolling(window).sum()
  # 相対力指数
  osc_rsi['rsi'] = osc_rsi['buy_power'] / (osc_rsi['buy_power']+osc_rsi['sell_power']) * 100
  return osc_rsi

# 指標計算関数を実行して結果をデータフレームに格納する
osc_rsi = OSC_RSI(df,14)
# 追加プロットのリストを作成する
osc_plot1 = [
    mpf.make_addplot((osc_rsi['rsi']), panel=2, ylabel='RSI', secondary_y=False)
]
```

```python
# 表題と凡例
ch_title = 'Stock Price Chart for ' + ticker
line_titles = ['RSI']
# チャート描画：指定可能なtypeは'ohlc'(default), 'candle', 'line', 'renko', 'pnf'
fig, axes = mpf.plot(df,type='candle',figsize =(16,9), style='yahoo',
         addplot=osc_plot1,volume=True,mav=(15,25,75), returnfig=True)
# 表題と凡例の指定
axes[0].set_title(ch_title)
axes[4].legend(line_titles)
```

【RSI解説】RSIが下方基準線(20-30%)を下回れば売られすぎ、上方基準線(70-80%)を上回れば買われすぎ。

RSIがボトムアウト後に下方基準線を上回る時に買い参入し、さらに上昇して上方基準線を上回る時に買いを手仕舞う。

## ストキャスティクス

【ストキャスティクス解説】

%Kと%Dを使う場合をファストストキャスティクス、%Dと%SD(Slow%D)を使う場合をスローストキャスティクスという。

下方基準線より下にあると売られすぎ、上方基準線より上にあると買われすぎと判断する。

```python
# Stochastic
def Stochastic(df, window, smooth_window):
    stochastic = pd.DataFrame()
    stochastic['%K'] = ((df['Close'] - df['Low'].rolling(window).min()) \
    / (df['High'].rolling(window).max() - df['Low'].rolling(window).min())) * 100
    stochastic['Hm'] = ((df['Close'] - df['Low'].rolling(window).min()).rolling(smooth_window).sum())
    stochastic['Lm'] = ((df['High'].rolling(window).max() - df['Low'].rolling(window).min()).rolling(smooth_window).sum())
    stochastic['%D'] = (stochastic['Hm'] / stochastic['Lm']) * 100
    stochastic['%SD'] = stochastic['%D'].rolling(smooth_window).mean()
    stochastic['UL'] = 80
    stochastic['DL'] = 20
    return stochastic

stochastic = Stochastic(df, 14, 3)

stochastic_plot = [
mpf.make_addplot((stochastic[['%K', '%D', '%SD', 'UL', 'DL']]), ylim=[0, 100], panel=2, ylabel='Stoch')
]
```

```python
# 表題と凡例
ch_title = 'Stock Price Chart for ' + ticker
line_titles = ['%K', '%D', '%SD', 'UL', 'DL']
# チャート描画：指定可能なtypeは'ohlc'(default), 'candle', 'line', 'renko', 'pnf'
fig, axes = mpf.plot(df,type='candle',figsize =(16,9), style='yahoo',
         addplot=stochastic_plot,volume=True, mav=(5,25,75), returnfig=True)
# 表題と凡例の指定
axes[0].set_title(ch_title)
axes[4].legend(line_titles)
```

【ストキャスティクス解説】

%Kと%Dを使う場合をファストストキャスティクス、%Dと%SD(Slow%D)を使う場合をスローストキャスティクスという。

下方基準線より下にあると売られすぎ、上方基準線より上にあると買われすぎと判断する。

## 究極のオシレーター (UO)

【UO解説】買い(売り)シグナルは、相場が安値(高値)を更新するなか、UOが最低値(最高値)を更新しないポジティブ(ネガティブ)・ダイバージェンスが起きた後、UOがダイバージェンス中の高値(安値)を超えた(割った)とき。

UOが70%を上回った時が利食いの目安、35%もしくは直近最低値を割ったときが損切の目安とされる。

```python
# subplotとして究極のオシレーター(UO)を描画する
# 各種指標の計算関数
def OSC_UO(df):
  osc_uo = pd.DataFrame()
  # 真の高値と真の安値
  osc_uo['th'] = np.maximum(df['High'], df['Close'].shift(1))
  osc_uo['tl'] = np.minimum(df['Low'], df['Close'].shift(1))
  # 買い圧力と全体の動き
  osc_uo['bp'] = df['Close'] - osc_uo['tl']
  osc_uo['ta'] = osc_uo['th'] - osc_uo['tl']
  # 短期・中期・長期の買い比率
  osc_uo['bshort'] = osc_uo['bp'].rolling(7).sum() / osc_uo['ta'].rolling(7).sum()
  osc_uo['bmiddle'] = osc_uo['bp'].rolling(14).sum() / osc_uo['ta'].rolling(14).sum()
  osc_uo['blong'] = osc_uo['bp'].rolling(28).sum() / osc_uo['ta'].rolling(28).sum()
  # 究極のオシレーター
  osc_uo['uo'] = (4*osc_uo['bshort']+2*osc_uo['bmiddle']+osc_uo['blong']) / 7 * 100
  return osc_uo

# 指標計算関数を実行して結果をデータフレームに格納する
osc_uo = OSC_UO(df)
# 追加プロットのリストを作成する
osc_plot1 = [
    mpf.make_addplot((osc_uo['uo']), panel=2, ylabel='UO', secondary_y=False)
]
```

```python
# 表題と凡例
ch_title = 'Stock Price Chart for ' + ticker
line_titles = ['UO']
# チャート描画：指定可能なtypeは'ohlc'(default), 'candle', 'line', 'renko', 'pnf'
fig, axes = mpf.plot(df,type='candle',figsize =(16,9), style='yahoo',
         addplot=osc_plot1,volume=True,mav=(15,25,75), returnfig=True)
# 表題と凡例の指定
axes[0].set_title(ch_title)
axes[4].legend(line_titles)
```

【UO解説】買い(売り)シグナルは、相場が安値(高値)を更新するなか、UOが最低値(最高値)を更新しないポジティブ(ネガティブ)・ダイバージェンスが起きた後、UOがダイバージェンス中の高値(安値)を超えた(割った)とき。

UOが70%を上回った時が利食いの目安、35%もしくは直近最低値を割ったときが損切の目安とされる。

## ウィリアムズ%R

【ウィリアムズ%R解説】10-20%(80-90%)程度に線をひき、それ以下(以上)の範囲を買われすぎ(売られすぎ)水準とする。

```python
# subplotとしてウィリアムズ%Rを描画する
# 各種指標の計算関数
def OSC_WPCTR(df,window):
  osc_wpctr = pd.DataFrame()
  # ウィリアムズ%R
  osc_wpctr['r1'] = df['High'].rolling(window).max() - df['Close']
  osc_wpctr['r2'] = df['High'].rolling(window).max() - df['Low'].rolling(window).min()
  osc_wpctr['wpctr'] = osc_wpctr['r1'] / osc_wpctr['r2'] * 100
  return osc_wpctr

# 指標計算関数を実行して結果をデータフレームに格納する
osc_wpctr = OSC_WPCTR(df,10)
# 追加プロットのリストを作成する
osc_plot1 = [
    mpf.make_addplot((osc_wpctr['wpctr']), panel=2, ylabel='WPCTR', secondary_y=False)
]
```

```python
# 表題と凡例
ch_title = 'Stock Price Chart for ' + ticker
line_titles = ['WPCTR']
# チャート描画：指定可能なtypeは'ohlc'(default), 'candle', 'line', 'renko', 'pnf'
fig, axes = mpf.plot(df,type='candle',figsize =(16,9), style='yahoo',
         addplot=osc_plot1,volume=True,mav=(15,25,75), returnfig=True)
# 表題と凡例の指定
axes[0].set_title(ch_title)
axes[4].legend(line_titles)
```

【ウィリアムズ%R解説】10-20%(80-90%)程度に線をひき、それ以下(以上)の範囲を買われすぎ(売られすぎ)水準とする。

## アルサー指数 (UI)

【UI解説】UIがゼロに近い場合、相場が上昇していることを示唆しており、UIが大きいほど相場の下落が厳しいことを示唆する。

平均的にUIの低い銘柄の方がリスクが低い。

```python
# subplotとしてアルサー指数(UI)を描画する
# 各種指標の計算関数
def OSC_UI(df,window):
  osc_ui = pd.DataFrame()
  # 期間最大からの比率
  osc_ui['dv'] = 100*(df['Close']/df['High'].rolling(window).max()-1)
  # 平方和の期間合計
  osc_ui['ssq'] = (osc_ui['dv'] * osc_ui['dv']).rolling(window).sum()
  # アルサー指数
  osc_ui['ui'] = np.sqrt(osc_ui['ssq']/window)
  return osc_ui

# 指標計算関数を実行して結果をデータフレームに格納する
osc_ui = OSC_UI(df,21)
# 追加プロットのリストを作成する
osc_plot1 = [
    mpf.make_addplot((osc_ui['ui']), panel=2, ylabel='UI', secondary_y=False)
]
```

```python
# 表題と凡例
ch_title = 'Stock Price Chart for ' + ticker
line_titles = ['UI']
# チャート描画：指定可能なtypeは'ohlc'(default), 'candle', 'line', 'renko', 'pnf'
fig, axes = mpf.plot(df,type='candle',figsize =(16,9), style='yahoo',
         addplot=osc_plot1,volume=True,mav=(15,25,75), returnfig=True)
# 表題と凡例の指定
axes[0].set_title(ch_title)
axes[4].legend(line_titles)
```

【UI解説】UIがゼロに近い場合、相場が上昇していることを示唆しており、UIが大きいほど相場の下落が厳しいことを示唆する。

平均的にUIの低い銘柄の方がリスクが低い。

## 順位相関指数 (RCI) (コード未検証）

```python
# subplotとして順位相関指数(RCI)を描画する
# 各種指標の計算関数
def rci(close: np.ndarray,
        timeperiod: int = 9) -> np.ndarray:
    rci = np.full_like(close, np.nan)
    rank_period = np.arange(1, timeperiod + 1)
    for i in range(timeperiod - 1, len(close)):
        rank_price = close[i - timeperiod + 1:i + 1]
        rank_price = np.argsort(np.argsort(rank_price)) + 1
        aa = 6 * sum((rank_period - rank_price)**2)
        bb = timeperiod * (timeperiod**2 - 1)
        rci[i] = (1 - aa / bb) * 100
    return rci
# 指標計算関数を実行して結果をリストに格納する
osc_rci = rci(df['Close'])
# 追加プロットのリストを作成する
osc_plot1 = [
    mpf.make_addplot((osc_rci), panel=2, ylabel='RCI', secondary_y=False)
]
```

```python
# 表題と凡例
ch_title = 'Stock Price Chart for ' + ticker
line_titles = ['RCI']
# チャート描画：指定可能なtypeは'ohlc'(default), 'candle', 'line', 'renko', 'pnf'
fig, axes = mpf.plot(df,type='candle',figsize =(16,9), style='yahoo',
         addplot=osc_plot1,volume=True,mav=(15,25,75), returnfig=True)
# 表題と凡例の指定
axes[0].set_title(ch_title)
axes[4].legend(line_titles)
```
