# mplfinanceライブラリを用いた株価チャートの描画

```python
import pandas as pd
from pandas_datareader import data as pdr
import mplfinance as fplt
```

## 株価データの取得

```python
from pandas_datareader import data as pdr
import datetime
import yfinance as yf 

start = "2023-4-1"
end = datetime.date.today()
ticker = "9983.T" # 9983:ファーストリテイリング
yf.pdr_override()
# yahooサイトからデータをダウンロード
df = pdr.get_data_yahoo(ticker, start, end)
df.head()
```

```python
# 指定可能なtype: 'ohlc'(default), 'candle', 'line', 'renko', 'pnf'
fplt.plot(df,type='renko',figsize =(16,5), title = "STOCK PRICE CHART FOR {0}".format(ticker))
```

```python
fplt.plot(df,type='renko',renko_params=dict(brick_size='atr', atr_length=2),style='yahoo',figsize =(18,7), 
          title = "RENKO CHART WITH ATR FOR {0}".format(ticker))
```
