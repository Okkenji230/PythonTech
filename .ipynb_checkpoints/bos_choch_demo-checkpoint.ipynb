ğŸ” BOSï¼ˆæ§‹é€ ã®ãƒ–ãƒ¬ã‚¤ã‚¯ï¼‰ã¨ CHoCHï¼ˆã‚­ãƒ£ãƒ©ã‚¯ã‚¿ãƒ¼ã®å¤‰åŒ–ï¼‰ã¨ã¯ï¼Ÿ
BOSï¼ˆBreak of Structureï¼‰
ã€€ç¾åœ¨ã®ãƒˆãƒ¬ãƒ³ãƒ‰ï¼ˆä¸Šæ˜‡ or ä¸‹é™ï¼‰ã®æ–¹å‘ã«ä¾¡æ ¼ãŒä»¥å‰ã®é«˜å€¤ã¾ãŸã¯å®‰å€¤ã‚’æ›´æ–°ï¼ˆçªç ´ï¼‰ã™ã‚‹ã“ã¨ã€‚
ã€€ğŸ‘‰ ã“ã‚Œã¯ãƒˆãƒ¬ãƒ³ãƒ‰ç¶™ç¶šã®ã‚µã‚¤ãƒ³ã¨ã•ã‚Œã¾ã™ã€‚

CHoCHï¼ˆChange of Characterï¼‰
ã€€ç¾åœ¨ã®ãƒˆãƒ¬ãƒ³ãƒ‰ã¨ã¯é€†æ–¹å‘ã«ä¾¡æ ¼ãŒãƒ–ãƒ¬ã‚¤ã‚¯ã—ãŸå ´åˆã€‚
ã€€ğŸ‘‰ ã“ã‚Œã¯ãƒˆãƒ¬ãƒ³ãƒ‰ã®åè»¢ã®å¯èƒ½æ€§ã‚’ç¤ºã™é‡è¦ãªã‚µã‚¤ãƒ³ã§ã™ã€‚

ğŸ”§ å…¨ä½“ã®æµã‚Œï¼ˆPythonå®Ÿè£…ï¼‰
â‘  ã‚¹ã‚¤ãƒ³ã‚°é«˜å€¤ãƒ»å®‰å€¤ã‚’æ¤œå‡ºã™ã‚‹
ä¸€å®šæœŸé–“å†…ã§ä¸€ç•ªé«˜ã„ã€Œã‚¹ã‚¤ãƒ³ã‚°ãƒã‚¤ã€ã¨ã€ä¸€ç•ªä½ã„ã€Œã‚¹ã‚¤ãƒ³ã‚°ãƒ­ãƒ¼ã€ã‚’è¦‹ã¤ã‘ã¾ã™ã€‚

```python
def detect_swings(df, lookback=3):
    df['swing_high'] = df['High'][
        (df['High'].shift(lookback) < df['High']) &
        (df['High'].shift(-lookback) < df['High'])
    ]
    df['swing_low'] = df['Low'][
        (df['Low'].shift(lookback) > df['Low']) &
        (df['Low'].shift(-lookback) > df['Low'])
    ]
    return df
```

â‘¡ BOSãƒ»CHoCHã‚’åˆ¤å®šã™ã‚‹
ä»¥ä¸‹ã®ã‚ˆã†ãªãƒ­ã‚¸ãƒƒã‚¯ã§é€²ã‚ã¾ã™ï¼š

å‰å›ã®ã‚¹ã‚¤ãƒ³ã‚°é«˜å€¤ãƒ»å®‰å€¤ã¨æ¯”è¼ƒ

ãƒˆãƒ¬ãƒ³ãƒ‰ãŒä¸Šæ˜‡ä¸­ã«å®‰å€¤ã‚’ä¸‹å›ã£ãŸã‚‰ã€ŒCHoCHã€

ãƒˆãƒ¬ãƒ³ãƒ‰ãŒä¸‹é™ä¸­ã«é«˜å€¤ã‚’ä¸Šå›ã£ãŸã‚‰ã€ŒCHoCHã€

åŒã˜æ–¹å‘ãªã‚‰ã€ŒBOSã€

```python
def find_bos_choch(df):
    structure = []
    trend = 'initial'
    last_high = -float('inf')
    last_low = float('inf')

    for i in range(len(df)):
        row = df.iloc[i]

        if not np.isnan(row['swing_high']):
            if (trend == 'down' or trend == 'initial') and row['High'] > last_high:
                structure.append('CHoCH')  # ä¸‹é™ãƒˆãƒ¬ãƒ³ãƒ‰ â†’ é«˜å€¤ã‚’è¶…ãˆã‚‹ â†’ è»¢æ›
                trend = 'up'
            elif (trend == 'up' or trend == 'initial') and row['High'] > last_high:
                structure.append('BOS')    # ä¸Šæ˜‡ãƒˆãƒ¬ãƒ³ãƒ‰ç¶™ç¶š
            else:
                structure.append(None)
            last_high = row['High']
        
        elif not np.isnan(row['swing_low']):
            if (trend == 'up' or trend == 'initial') and row['Low'] < last_low:
                structure.append('CHoCH')  # ä¸Šæ˜‡ãƒˆãƒ¬ãƒ³ãƒ‰ â†’ å®‰å€¤ã‚’å‰²ã‚‹ â†’ è»¢æ›
                trend = 'down'
            elif (trend == 'down' or trend == 'initial') and row['Low'] < last_low:
                structure.append('BOS')    # ä¸‹é™ãƒˆãƒ¬ãƒ³ãƒ‰ç¶™ç¶š
            else:
                structure.append(None)
            last_low = row['Low']

        else:
            structure.append(None)
    df['structure'] = structure
    return df
```

âœ… å®Ÿéš›ã«ä½¿ã£ã¦ã¿ã‚‹

```python
# è©•ä¾¡ã™ã‚‹éŠ˜æŸ„
ticker = '4755.T'
# æœŸé–“ã®æŒ‡å®š
end_date = datetime.today()
start_date = end_date - timedelta(days=730)

# æ ªä¾¡ãƒ‡ãƒ¼ã‚¿å–å¾—
df = yf.download(ticker, start=start_date, end=end_date, interval='1d', multi_level_index=False)
```

```python
# ã‚¹ã‚¤ãƒ³ã‚°ãƒã‚¤ãƒ³ãƒˆ â†’ BOS/CHoCHæ¤œå‡º
df = detect_swings(df,lookback=5)
df = find_bos_choch(df)

# çµæœã®è¡¨ç¤ºï¼ˆBOSã¨CHoCHã®ã¿ï¼‰
print(df[df['structure'].notna()][['High', 'Low', 'structure']])
```

ğŸ“ˆ ã‚ªãƒ—ã‚·ãƒ§ãƒ³ï¼šãƒãƒ£ãƒ¼ãƒˆã§å¯è¦–åŒ–

```python
plt.figure(figsize=(14, 6))
plt.plot(df['Close'], label='Close Price')

# BOS/CHoCHã«ãƒãƒ¼ã‚«ãƒ¼ã‚’è¡¨ç¤º
for idx, row in df[df['structure'].notna()].iterrows():
    color = 'green' if row['structure'] == 'BOS' else 'red'
    plt.scatter(idx, df.loc[idx, 'Close'], color=color, label=row['structure'])

plt.title('Detecting BOS(Cont,G) and CHoCH(Rev,R)')
plt.show()
```
