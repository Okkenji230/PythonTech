# Pythonによるログデータの分析小技集

### Kaggleにあるログデータ(https://www.kaggle.com/mehulkatara/windows-event-log) を読み込む

```python
import pandas as pd
df1 = pd.read_parquet('eventlog.parquet')
df1.reset_index(inplace=True)
df1.head()
```

```python
df1.info()
```

## 特定の値を持つ行に限定する

```python
ch_machine = 'LAPTOP-1MKMTVPM'
df2 = df1[df1['MachineName'].str.contains(ch_machine)]
df2.head()
```

# 丸め時刻毎に定型テキストの出現頻度を集計してグラフ化する

### 日単位の丸め時間変数を作成する

```python
df21=df2.assign(time_round=df2['timestamp'].dt.floor('D'))
df21.head()
```

### 日とEntryType毎に行数をカウントする

```python
df3 = df21.groupby(['time_round','EntryType']).count()
df31 = df3.reset_index()
df32 = df31.set_index('time_round')
df32.head()
```

### 各EntryTypeを列とする変数を作成する

```python
df4 = df32.pivot(columns='EntryType',values='MachineName')
df41 = df4.fillna(0) # 欠損値にゼロを代入する
df41.head()
```

### エラーの発生の推移をグラフ化する

```python
import matplotlib.pyplot as plt 
from matplotlib import rcParams; rcParams['font.family']='MS Gothic' 
df41['Warning'].plot(figsize=(9,6))
df41['Error'].plot(figsize=(9,6))
plt.title('日別の警告とエラーの数   マシン: '+ch_machine)
plt.legend()
plt.show()
# plt.savefig('event_log_messages.png')
```

## 時間範囲を指定して検索する

```python
# 時刻（範囲）を指定したデータの検索
import datetime
from datetime import timedelta
# 日付、時、時間範囲を指定する
query_date=20201023
query_time_h=8
query_hours=21
# 日付関連値を求める
c_year=query_date//10000
c_month=(query_date//100)%100
c_day=query_date%100
# データ開始時刻の日付値
dt1=datetime.datetime(c_year,c_month,c_day,query_time_h,0,0)
# データ終了時刻の日付値
dt2 = dt1 + timedelta(hours=query_hours)
 # 指定した時間範囲内のデータに絞り込む
df52 = df2.query('timestamp >= @dt1 & timestamp <= @dt2')
df52.head()
```

## 定型テキストの出現傾向を表に要約する

```python
import textwrap
# strftime()で日本語を使用するのに必要
import locale
locale.setlocale(locale.LC_ALL, '')
```

### 月の上旬・中旬・下旬を判定する

```python
def func_part_of_month(tround):
    '''月の上旬・中旬・下旬の判定関数'''
    if tround.day <=10 :
        return '上旬'
    elif tround.day <= 20 :
        return '中旬'
    else :
        return '下旬'
```

```python
# 年月を表す変数を導入
df62=df2.assign(yr_month=df2['timestamp'].dt.strftime('%Y年%m月'))
# 上/中/下旬を表す変数を導入
df62=df62.assign(part_of_month=df62['timestamp'].apply(func_part_of_month))
df62.sample(5)
```

### グループ毎に最頻出の文字列を求める

```python
df63=df62.groupby(['yr_month','part_of_month'])['Message'].apply(lambda x:x.value_counts().index[0])
df64 = pd.DataFrame(df63)
df64.head()
```

### 長い文字を切り詰め、途中に改行を入れる

```python
def func_wrap(s):
    '''長い文字を切り詰め、途中に改行を入れる関数'''
    s_trunc = textwrap.shorten(s,100)
    s_wrap_list = textwrap.wrap(s_trunc, 25)
    return '\n'.join(s_wrap_list)
```

```python
# 文字列切り詰めと改行挿入
df64['f_msg'] = df64['Message'].apply(func_wrap)
df65 = df64.reset_index()
df66 = df65.set_index('yr_month')
df66.head()
```

### 月を行、上中下旬を列とする要約表を作成する

```python
# 各part_of_monthを列とする変数を作成する
df67 = df66.pivot(columns='part_of_month',values='f_msg')
# 欠損値の表示方法の指定
df68 = df67.fillna('記録なし')
# データの行数に応じて表の高さを決める
tbl_height = len(df68) * 1.0
# 図形の描画
fig=plt.figure(figsize=(8,tbl_height))
ax=fig.add_subplot(111); ax.axis('off')
# matplotlibのtable関数で表を描く
tbl=ax.table(cellText=df68.values,bbox=[0,0,1,1],colLabels=df68.columns,rowLabels=df68.index)
plt.plot()
```
