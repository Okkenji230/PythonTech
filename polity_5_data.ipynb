# Polity 国際政治データ

```python
import pandas as pd
import numpy as np
df1=pd.read_excel('p5v2018.xlsx',sheet_name='p5v2018')
df1.head()
```

```python
df1.info()
```

```python
df2=df1.loc[:,['country','year','democ','autoc','polity','polity2']]
df2.head()
```

```python
df2.describe()
```

```python
# matplotlibライブラリのインポート
import matplotlib.pyplot as plt
# matplotlibで日本語を使えるようにする（windowsの場合）
from matplotlib import rcParams
rcParams['font.family']='MS Gothic'
# 描画の準備
fig=plt.figure(figsize=(20,5))
ax1=fig.add_subplot(141)
ax1.set_title('democ')
ax1.hist(df2['democ'])
ax2=fig.add_subplot(142)
ax2.set_title('autoc')
ax2.hist(df2['autoc'])
ax3=fig.add_subplot(143)
ax3.set_title('polity')
ax3.hist(df2['polity'])
ax4=fig.add_subplot(144)
ax4.set_title('polity2')
ax4.hist(df2['polity2'])
```

```python
# -30より小さい値は欠損値とする
df3 = df2.copy()
df3.loc[df3['democ'] < -30, 'democ'] = np.nan
df3.loc[df3['autoc'] < -30, 'autoc'] = np.nan
df3.loc[df3['polity'] < -30, 'polity'] = np.nan
df3.loc[df3['polity2'] < -30, 'polity2'] = np.nan
```

```python
# 描画の準備
fig=plt.figure(figsize=(20,5))
ax1=fig.add_subplot(141)
ax1.set_title('democ')
ax1.hist(df3['democ'])
ax2=fig.add_subplot(142)
ax2.set_title('autoc')
ax2.hist(df3['autoc'])
ax3=fig.add_subplot(143)
ax3.set_title('polity')
ax3.hist(df3['polity'])
ax4=fig.add_subplot(144)
ax4.set_title('polity2')
ax4.hist(df3['polity2'])
```

```python
# pickleファイルに出力
df3.to_pickle('polity.pkl')
```
