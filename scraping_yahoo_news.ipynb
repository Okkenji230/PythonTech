```python
# ライブラリのインポート
import requests
from bs4 import BeautifulSoup
import re
import japanize_matplotlib # matplotlib日本語
from google.colab import files # 画像ファイルダウンロード用
```

requests.get()を利用して、Yahoo newsサイトの情報をダウンロードする

```python
url = 'https://www.yahoo.co.jp'
res = requests.get(url)
```

BeautifulSoupに先程取得したウェブサイトの情報とパーサー'html.parser'を渡す

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

find_all()に引数'a'を渡して検索してみる

```python
elems = soup.find_all('a')
elems
```

ヘッドラインニュースには'news.yahoo.co.jp/pickup'が含まれてるのでreライブラリを用いて抽出する

```python
elems = soup.find_all(href=re.compile('news.yahoo.co.jp/pickup'))
elems
```

contentsを繰り返し適用して一つ目のニュースタイトルを取り出す

```python
elems[0].contents[0].contents[0].contents[0].contents[0].contents[0]
```

今度はリンクのurlだけを取得する

```python
elems[0].attrs['href']
```

以上を全ての要素について表示させる

```python
for elem in elems:
  print(elem.contents[0].contents[0].contents[0].contents[0].contents[0])
  print(elem.attrs['href'])
```
