## MACDの描画

(参考記事)
https://plainenglish.io/blog/plot-stock-chart-using-mplfinance-in-python-9286fc69689

```python
start = "2023-4-1"
end = datetime.date.today()
ticker = "4755.T" # 4755は楽天グループ
yf.pdr_override()
# yahooサイトからデータをダウンロード
df3 = pdr.get_data_yahoo(ticker, start, end)
```

```python
# MACDの諸指標を計算する関数
# .ewm(span=n)のとき指数平滑係数α=2/(span+1) ただしspan>=1
def MACD(df, window_slow, window_fast, window_signal):
    macd = pd.DataFrame()
    macd['ema_slow'] = df['Close'].ewm(span=window_slow).mean()
    macd['ema_fast'] = df['Close'].ewm(span=window_fast).mean()
    macd['macd'] = macd['ema_slow'] - macd['ema_fast']
    macd['signal'] = macd['macd'].ewm(span=window_signal).mean()
    macd['diff'] = macd['macd'] - macd['signal']
    macd['bar_positive'] = macd['diff'].map(lambda x: x if x > 0 else 0)
    macd['bar_negative'] = macd['diff'].map(lambda x: x if x < 0 else 0)
    return macd

# MACD計算関数を呼び出し、結果をデータフレームに格納する
macd = MACD(df3, 12, 26, 9)

# mplfinanceのmake_addplotメソッドを用いてチャートに付け加えるプロットを作成する
macd_plot = [
    mpf.make_addplot((macd['macd']), color='#D2691E', panel=2, ylabel='MACD', secondary_y=False), #D2691Eはチョコレート色
    mpf.make_addplot((macd['signal']), color='#800080', panel=2, secondary_y=False), #800080は紫
    mpf.make_addplot((macd['bar_positive']), type='bar', color='#4dc790', panel=2),
    mpf.make_addplot((macd['bar_negative']), type='bar', color='#fd6b6c', panel=2),
]
```

カラーコード一覧：https://itsakura.com/html-color-codes

チョコレート色で、株価上昇時に先に反応して上向くのがMACD。紫色で後から上向くのがMACDシグナル(移動平均)。棒グラフは両者の差(MACDヒストグラム)

```python
# 上記で作成したプロットのリストを引数addplot=で指定してチャートの下にMACDのプロットを付け加える
mpf.plot(df3,type='candle',figsize =(12,6), style='yahoo',
          volume=True, addplot=macd_plot,
          title = "STOCK PRICE CHART FOR {0}".format(ticker))
```

## ストキャスティクスの描画

```python
# ストキャスティクス(オシレーター)の諸指標を計算する関数
def Stochastic(df, window, smooth_window):
    stochastic = pd.DataFrame()
    stochastic['%K'] = ((df['Close'] - df['Low'].rolling(window).min()) \
    / (df['High'].rolling(window).max() - df['Low'].rolling(window).min())) * 100
    stochastic['Hm'] = ((df['Close'] - df['Low'].rolling(window).min()).rolling(smooth_window).sum())
    stochastic['Lm'] = ((df['High'].rolling(window).max() - df['Low'].rolling(window).min()).rolling(smooth_window).sum())
    stochastic['%D'] = (stochastic['Hm'] / stochastic['Lm']) * 100
    stochastic['%SD'] = stochastic['%D'].rolling(smooth_window).mean()
    stochastic['UL'] = 80
    stochastic['DL'] = 20
    return stochastic

# 上記関数を window=14及びsmoothing_window=3として呼び出し、結果をデータフレームに格納する
stochastic = Stochastic(df3, 14, 3)

# mplfinanceのmake_addplotメソッドを用いてチャートに付け加えるプロットを作成する
stochastic_plot = [
mpf.make_addplot((stochastic[['%K', '%D', '%SD', 'UL', 'DL']]), ylim=[0, 100], panel=2, ylabel='Stoch')
]
```

```python
# 上記で作成したプロットのリストを引数addplot=で指定してチャートの下にストキャスティクスのプロットを付け加える
mpf.plot(df3, type='candle',figsize =(12,6), volume=True, addplot=stochastic_plot, style='yahoo',
          title = "STOCK PRICE CHART FOR {0}".format(ticker))
```

## 一つの図にまとめて表示させる

```python
macd = MACD(df3, 12, 26, 9)
stochastic = Stochastic(df3, 14, 3)
plots = [
mpf.make_addplot((macd['macd']), color='#606060', panel=2, ylabel='MACD (12,26,9)', secondary_y=False),
mpf.make_addplot((macd['signal']), color='#1f77b4', panel=2, secondary_y=False),
mpf.make_addplot((macd['bar_positive']), type='bar', color='#4dc790', panel=2),
mpf.make_addplot((macd['bar_negative']), type='bar', color='#fd6b6c', panel=2),
mpf.make_addplot((stochastic[['%D', '%SD', 'UL', 'DL']]), ylim=[0, 100], panel=3, ylabel='Stoch (14,3)')
]

mpf.plot(df3, type='candle', style='yahoo', mav=(5,20), volume=True,
          addplot=plots, panel_ratios=(3,1,3,3), figscale=1.5,
          title = "STOCK PRICE CHART FOR {0}".format(ticker))
```
